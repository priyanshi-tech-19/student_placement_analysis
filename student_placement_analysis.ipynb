```python
print(df['mock_interview_score'].isnull().sum())
print(df['GitHub_repos'].isnull().sum())
```

```python
#Fill missing values in GitHub_repos and mock_interview_score with their median. Confirm no nulls remain.

df['GitHub_repos'] = df['GitHub_repos'].fillna(df['GitHub_repos'].median())
#df['mock_interview_score'] = pd.to_numeric(df['mock_interview_score'],errors='coerce')
df['mock_interview_score'] = df['mock_interview_score'].fillna(df['mock_interview_score'].median())
```

```python
print(df['mock_interview_score'].isnull().sum())
print(df['GitHub_repos'].isnull().sum())
```

```python
#Find the 25th, 50th, and 75th percentile of salary_lpa using NumPy.
np.percentile(df['salary_lpa'].dropna(),[25,50,75])
```

```python
#Create a new column high_cgpa: 1 if cgpa > 8, else 0. Use np.where()
df [ 'high_cgpa'] = np.where(df['cgpa'] > 8,1,0)
```
